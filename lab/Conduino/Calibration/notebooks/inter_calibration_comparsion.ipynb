### How much do calibrations change?

Comparing calibrations Mar 29, May 02, July 6, July 12, Aug 09, Aug 17

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import scipy.io as sio
import sys
sys.path.append('/ocean/kramosmu/MultipleCanyons/lab/Conduino/scripts')
import conduino_tools as ctools
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
dummy = np.linspace(0,4.2,50)

A1_29mar = ctools.densANK1_29Mar19(dummy)
A1_02may = ctools.densANK1_02May19(dummy)
A1_06jul = ctools.densANK1_06Jul19(dummy)
A1_12jul = ctools.densANK1_12Jul19(dummy)
A1_09aug = ctools.densANK1_09Aug19(dummy)
A1_17aug = ctools.densANK1_17Aug19(dummy)

A2_29mar = ctools.densANK2_29Mar19(dummy)
A2_02may = ctools.densANK2_02May19(dummy)
A2_06jul = ctools.densANK2_06Jul19(dummy)
A2_12jul = ctools.densANK2_12Jul19(dummy)
A2_09aug = ctools.densANK2_09Aug19(dummy)
A2_17aug = ctools.densANK2_17Aug19(dummy)

P_29mar = ctools.densP_29Mar19(dummy)
P_02may = ctools.densP_02May19(dummy)
P_06jul = ctools.densP_06Jul19(dummy)
P_12jul = ctools.densP_12Jul19(dummy)
P_09aug = ctools.densP_09Aug19(dummy)
P_17aug = ctools.densP_17Aug19(dummy)

labels = ['29 Mar', '02 May', '06 Jul', '12 Jul (independent)', '09 Aug','17 Aug']
densitiesA1 = [A1_29mar, A1_02may, A1_06jul, A1_12jul, A1_09aug, A1_17aug]
densitiesA2 = [A2_29mar, A2_02may, A2_06jul, A2_12jul, A2_09aug, A2_17aug]
densitiesP = [P_29mar, P_02may, P_06jul, P_12jul, P_09aug, P_17aug]
```

```python
fig, ax = plt.subplots(1,3, figsize=(10,3))
for densA1, densA2, densP, label in zip(densitiesA1, densitiesA2, densitiesP, labels):
    ax[0].plot(dummy, densA1, label=label)
    ax[1].plot(dummy, densA2, label=label)
    ax[2].plot(dummy, densP, label=label)
   
ax[2].legend()
ax[0].set_xlabel(r'reading $\phi$')
ax[1].set_xlabel(r'reading $\phi$')
ax[2].set_xlabel(r'reading $\phi$')
ax[0].set_ylabel(r'density $\rho$ / g/cm$^3$')
ax[0].set_title('upstream ANK1')
ax[1].set_title('downstream ANK2')
ax[2].set_title('profiler P20')
```

These are the last calibrations I've done for all 3 probes. Calibrations for ANK1 look very consisten compared to the other 2 probes.

```python
fig, ax = plt.subplots(1,3, figsize=(10,3), sharey=True)
for densA1, densA2, densP, label in zip(densitiesA1, densitiesA2, densitiesP, labels):
    ax[0].plot(dummy[1:], (densA1[1:]-densA1[:-1])/(dummy[1:]-dummy[:-1]), label=label)
    ax[1].plot(dummy[1:], (densA2[1:]-densA2[:-1])/(dummy[1:]-dummy[:-1]), label=label)
    ax[2].plot(dummy[1:], (densP[1:]-densP[:-1])/(dummy[1:]-dummy[:-1]), label=label)
   
ax[2].legend()
ax[0].set_xlabel('reading')
ax[1].set_xlabel('reading')
ax[2].set_xlabel('reading')
ax[0].set_ylabel(r'd$\rho$/d$\phi$')
ax[0].set_title('upstream ANK1')
ax[1].set_title('downstream ANK2')
ax[2].set_title('profiler P20')
```

The error associated to having slightly larger or smaller readings than the reading $r_0$ at the time of calibration has to do with the derivative of the calibration curve.
