### Plot experiment $\Delta \rho$ and recovery to f0

```python
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from numpy.polynomial import polynomial as P
import os
import pandas as pd
#import seaborn as sns
import scipy.io as sio
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

import sys
sys.path.append('/ocean/kramosmu/MultipleCanyons/lab/Conduino/scripts')
import profiler_tools as ptools
import conduino_tools as ctools
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
def plot_uprofile(ax, col='0.8'):
    ax.axvline((15+21.0)/60, color=col, linestyle=':')
    ax.axvline((15+71.0)/60, color=col, linestyle=':')
        
```

```python
# Experiment plus recovery
fig, (ax0,ax1) = plt.subplots(2,1, figsize=(10,5), sharex=True)

labels = ['05-experiment','06-recovery','09-experiment','10-recovery']
time_offset = [0,7*60,0,7*60]
filepaths = ['../experiments/exp_17aug19/05_experiment1.mat',
             '../experiments/exp_17aug19/06_recovery1.mat',
             '../experiments/exp_17aug19/09_experiment2.mat',
             '../experiments/exp_17aug19/10_recovery2.mat',
             ]
sns.set_palette('plasma',n_colors=len(filepaths))
for filepath, lab, offset, ii in zip(filepaths, labels, time_offset, range(len(filepaths))):
    mat = sio.loadmat(filepath, squeeze_me=True)
    P1 = mat['ch1']
    P2 = mat['ch2']
    time = mat['time']
    dens1 = ctools.densANK1_17Aug19(P1)
    dens2 = ctools.densANK2_17Aug19(P2)
    if offset == 0:
        ini_dens1 = dens1[0]
        ini_dens2 = dens2[0]
        
    else:
        mat0 = sio.loadmat(filepaths[ii-1], squeeze_me=True)
        P1_0 = mat0['ch1']
        P2_0 = mat0['ch2']
        dens1_0 = ctools.densANK1_17Aug19(P1_0)
        dens2_0 = ctools.densANK2_17Aug19(P2_0)
        ini_dens1 = dens1_0[0]
        ini_dens2 = dens2_0[0]
    
    ax0.plot(offset+time,dens1-ini_dens1, label=lab)
    ax1.plot(offset+time,dens2-ini_dens2, label=lab)
   
ax0.legend(bbox_to_anchor=(1,1))
ax1.legend(bbox_to_anchor=(1,1))

ax1.set_ylabel(r'$\Delta \rho$ / g cm$^{-3}$')
ax0.set_ylabel(r'$\Delta \rho$ / g cm$^{-3}$')
ax0.set_title('ANK1 - upstream probe')
ax1.set_title('ANK2 - downstream probe')
ax1.set_xlabel('time / s')
```
