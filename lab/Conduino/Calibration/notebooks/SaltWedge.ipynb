Probes A and K near surface; probes 37 and 38 closer to bottom.

|Channel|Probe|
|::|::|
|'ch1':| 37|
|'ch2': |K|
|'ch3': |A|
|'ch4': |38|

![first probes](salt_wedge_set_up.png)


```python
def dens_K(read):
    rho_K = 0.997645 + 0.004937*read + 0.006870*(read**2) - 0.000332*(read**3)
    return(rho_K)

def dens_A(read):
    rho_A = 0.997590 + 0.006686*read + 0.001903*(read**2) + 0.000142*(read**3)
    return(rho_A)
```

### Salt wedge 01

```python
fig, (ax0) = plt.subplots(1,1, figsize=(10,5), sharey=True)

labels = ['37 - deep, fresher','K - shallow, fresher','A - shallow, salty','38 - deep, salty']
colors = ['green','limegreen','rebeccapurple', 'mediumpurple']

filepath = '../SaltWedge/SaltWedge01.mat'
print(filepath)
    
mat = sio.loadmat(filepath, squeeze_me=True)
P37 = mat['ch1']
PK = mat['ch2']
PA = mat['ch3']
P38 = mat['ch4']

time = mat['time']
ax0.plot(time,P37,'-' ,label=labels[0], color=colors[0])
ax0.plot(time,PK,'-', label=labels[1], color=colors[1])
ax0.plot(time,PA,'-', label=labels[2], color=colors[2])
ax0.plot(time,P38,'-', label=labels[3], color=colors[3])

ax0.legend(bbox_to_anchor=(1,1))
ax0.set_ylabel('reading / au')
ax0.set_xlabel('time / s')
```

```python
fig, (ax0) = plt.subplots(1,1, figsize=(10,5), sharey=True)

ax0.plot(time,dens_K(PK),'-', label=labels[1], color=colors[1])
ax0.plot(time,dens_A(PA),'-', label=labels[2], color=colors[2])

ax0.legend(bbox_to_anchor=(1,1))
ax0.set_ylabel('density / g cm$^{-3}$')
ax0.set_xlabel('time / s')
```
